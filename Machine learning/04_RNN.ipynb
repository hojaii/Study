# 중심 극한정리

실험을 충분히 많이하면 나오는 분포가 정규분포의 형태를 지난다는 이론

```python
def bernouli_trial(p): # 확률 p보다 적을때 1, 그렇지 않을 때 1 (0 <= p <=1)
  return 1 if np.random.random()<p else 0
```

```python
cnt = 0
for _ in range(1000):
  re = bernouli_trial(1/6)
  print(re, end=",") # 주사위를 던져서 1이 나왔는지 출력 (1: 나왔다, 0: 안나왔다.)
  if re == 1:
    cnt += 1 # 주사위를 던져서 1면이 나온 횟수를 증가
print()
print(cnt)
```

```python
def bernouli_trials(n, p): # n번 bernouli_trials을 수정 했을 때 사건이 발생할 횟수를 구하는 함수
  return sum(bernouli_trial(p)for _ in range(n))
```

```python
re = bernouli_trials(1000, 1/6)
print(re/1000)
```

```python
for i in range(100):
  print(bernouli_trials(1000, 1/6), end = ",")
  if i%20 == 19:
    print()
```

```python
from collections import Counter
import math
import scipy.stats as st
```

```python
data = [bernouli_trials(1000,1/6) for _ in range(100)]
histogram = Counter(data) #분포수 세기
print(histogram.values())
print(histogram.keys())
print([v/100 for v in histogram.values()])
mu = 1/6*1000
print(f"기대값:{mu}") #기대값(평균)
sigma = math.sqrt(1000 * 1/6*(1-1/6))
print("표준편차:",sigma)#표준편차
xs = range(min(data),max(data)+1)
ys = [st.norm.cdf(i+0.5, mu, sigma) - st.norm.cdf(i-0.5, mu, sigma) for i in xs]
ys2 =[st.norm.pdf(i, mu, sigma) for i in xs]
print(ys)
print(ys2)
plt.bar(histogram.keys(),[v/100 for v in histogram.values()], color='r')
plt.plot(xs,ys2)
plt.show()
```

```python
def bernouli_histogram(p,n,nps):#p:확률, n:시도 횟수, nps:bernouli_trials(n,p)를 호출할 횟수
  data = [bernouli_trials(n,p) for _ in range(nps)]
  histogram = Counter(data)  
  mu = p*n
  sigma = math.sqrt(n * p*(1-p))
  xs = range(min(data),max(data)+1)  
  ys2 =[st.norm.pdf(i, mu, sigma) for i in xs]  
  plt.bar(histogram.keys(),[v/nps for v in histogram.values()], color='r')
  plt.plot(xs,ys2)
  plt.show()
```

```python
bernouli_histogram(1/6, 100, 1000) # 중심극한 정리
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
train_seq = pad_sequences(train_input, maxlen=100)
val_seq = pad_sequences(val_input, maxlen=100)
test_seq = pad_sequences(test_input, maxlen=100)
```

```python
from tensorflow import keras
model = keras.Sequential()
model.add(keras.layers.SimpleRNN(8, input_shape=(100, 500)))
model.add(keras.layers.Dense(1, activation = "sigmoid"))
```

```python
train_oh = keras.utils.to_categorical(train_seq)
train_oh.shape
```

```python
np.sum(train_oh[0][0])
```

```python
val_oh = keras.utils.to_categorical(val_seq)
test_oh = keras.utils.to_categorical(test_seq)
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics = ["accuracy"])
cp = keras.callbacks.ModelCheckpoint("best-simplernn-model.h5", save_best_only=True)
es = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
history = model.fit(train_oh, train_target, epochs=100, batch_size=64, validation_data=(val_oh,val_target), callbacks=[cp, es])
```

```python
import matplotlib.pyplot as plot
fig, loss_ax = plt.subplots(figsize =(6,3))
acc_ax = loss_ax.twinx()

loss_ax.plot(history.history["loss"], "y", label = "train loss")
loss_ax.plot(history.history["val_loss"], "g", label = "val_loss")
loss_ax.set_xlabel("epoch")
loss_ax.set_ylabel("loss")

acc_ax.plot(history.history["accuracy"], "b", label="train acc")
acc_ax.plot(history.history["val_accuracy"], "r", label="val acc")
acc_ax.set_ylabel("accuracy")

loss_ax.legend(loc="upper left")
acc_ax.legend(loc="lower left")
plt.show()
```
