```python
# 상자 그림(BoxPlot)
boxplot = plt.figure()
boxplot_axe = boxplot.add_subplot(1, 1, 1)
boxplot_axe.boxplot([mtcars["mpg"], mtcars["wt"]])
plt.show()
```

```python
# 극단치 구하기
# 4분위 수 구하기
first_quartile = mtcars["mpg"].quantile(0.25) # 25% 지점의 수 -> 10
third_quartile = mtcars["mpg"].quantile(0.75) # 75% 지점의 수 -> 30
print(first_quartile, third_quartile)
iqr = third_quartile - first_quartile # 30 - 10 -> IQR
print("IQR", iqr)
```

```python
# 극단치 경계 구하기
# 상단 극단치 경계 : 3Q +1.5 * IQR
border_top = third_quartile + 1.5 * iqr
print("mpg의 상단 극단치 경계:", border_top)
```

```python
# 하단 극단치 경계 : 1Q - 1.5 * IQR
border_bottom = first_quartile - 1.5 * iqr
print("mpg의 하단 극단치경계:", border_bottom)
```

```python
# mpg값이 상단 극단치를 넘어간 차량은 무엇?
mpg_outliers = mtcars[mtcars["mpg"] > border_top]
mpg_outliers
```

```python
# 파이 차트
# 100 에서 부분의 비율율 표시
data = [4, 12, 11, 3, 1]
labels = ["A", "B", "C", "D", "E"]

fig = plt.figure()
pie_axe = fig.add_subplot(1, 1, 1)

pie_axe.pie(data, #데이터
            labels=labels,
           autopct="%.2f%%", # 퍼센트표시
            shadow=True, # 그림자 표시 여부
            startangle=90, # 회전 시작 각도
            explode=(0.2, 0.1, 0, 0, 0), # 튀어나오는 정도
            colors=("red", "lightblue", "green", "yellow", "gray")
           )
plt.show()
```

```python
# 2012년 미국 주별 실업률 데이터
us_unemp = "./US_Unemployment_Oct2012.csv"
state_data = pd.read_csv(us_unemp)
state_data.head()
```

```python
# 지도에 연결할 지역 데이터
import json
state_json = json.load(open("./folium_us-states.json"))
state_json
```

```python
map = folium.Map(location=[40,-90], #위도, 경도
                zoom_start=3) #시작확대 비율
# 지리정보 매핑
folium.Choropleth(geo_data = state_json, # 지리정보
                 data = state_data, # 실업률 정보
                 columns=["State", "Unemployment"],
                 key_on="feature.id",
                  fill_color="YlGn",
                 legend_name="Unemployment Rate (%)"
                 ).add_to(map)
map
```
