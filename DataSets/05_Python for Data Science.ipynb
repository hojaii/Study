```python
# 슬라이싱 구문을 이용한 데이터 추출
# 전체 행, year, pop컬럼 선택
#[행범위, 열범위]
subset2 = df.loc[:, ["year", "pop"]]
subset2
```

```python
# range를 활용한 데이터 추출
u3_range = range(4) # 0 ~ 3
print(u3_range)

df.iloc[:, u3_range].head() # 전체형, 0 ~ 3 컬럼
```

```python
# 짝수번째 컬럼만 출력
subset4 = df.iloc[:, range(0, df.shape[1], 2)]
subset4.head()
```

```python
# df을 year 컬럼으로 그룹핑
grouped_year_df = df.groupby("year") # year 컬럼으로 그룹핑
print(type(grouped_year_df))
grouped_year_df
```

```python
# year 정보로 그룹핑된 데이터 프레임에서 lifeExp의 평균
grouped_year_df["lifeExp"].mean()
```

```python
# 데이터 프레임에서 활용 가능한 기초 통계량
grouped_year_df.describe()
# 데이터 타입이 object인것은 통계를 잡을수 없음
```

```python
# 여러 컬럼의 산술 평균
# 년도별 기대수명, 인구당 GDP의 평균
grouped_year_df [["lifeExp", "gdpPercap"]].mean()
```

```python
# 빈도 수 세기
df.groupby("continent")["country"].nunique()
```

```python
# 시각화
# 연도별 기대 수명의 평균 데이터 프레임
life_exp = df.groupby("year")["lifeExp"].mean()
life_exp
```

```python
# 차트로 그리기
%matplotlib inline
life_exp.plot()
```
